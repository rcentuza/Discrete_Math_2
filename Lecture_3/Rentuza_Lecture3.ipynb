## Graph <br>
A graph denoted as $G = (V,E)$ consist of a non-empty set of vertices or nodes $V$ and a set of edges $E$ <br>

<b>Example:</b>

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
G.add_node('D')
G.add_edges_from([('A','B'),('C','A'),('B','C')])

pos = nx.planar_layout(G)

nx.draw_networkx_nodes(G,pos, node_size=500)
nx.draw_networkx_edges(G,pos)
nx.draw_networkx_labels(G,pos)
plt.show()
```

$\textbf{Vertex Set} = \{v_A,v_B,v_C\}$ <br>
$\textbf{Edge Set} = \{e_x,e_y,e_z\}$

$\textbf{Edge-Endpoint Function}$
\begin{gather}
\begin{bmatrix}
\begin{array}{c|c}
Edge & Endpoints \\
e_x & \{v_A,v_B\} \\
e_y & \{v_A,v_C\} \\
e_z & \{v_B,v_C\} \\
\end{array}
\end{bmatrix}
\end{gather}

$Incident:$ <br>

$e_x$ is incident to $\{v_A,v_B\}$ <br>
$e_y$ is incident to $\{v_A,v_C\}$ <br>
$e_z$ is incident to $\{v_B,v_C\}$ <br>

$Adjacent:$ <br>

$v_A$ is adjacent to $\{v_B,v_C\}$ <br>
$v_B$ is adjacent to $\{v_A,v_C\}$ <br>
$v_C$ is adjacent to $\{v_A,v_B\}$ <br>

$Isolated:$ <br>

$v_D$ is isolated <br>

## Complete Graphs($K_n$)

  $\forall \, a,b \in V \, and \, a \neq b, \{a,b\} \in E$ <br>

  <b>Example:</b>

  $K_1 , K_2, K_3$

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
G.add_edges_from([('A','B'),('A','C'),('B','C'),('X','Y')])  # Corrected edge specification
G.add_node ('Z')

pos = nx.planar_layout(G)

nx.draw_networkx_nodes(G, pos, node_size=500)
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color='black')
nx.draw_networkx_labels(G, pos)
plt.show()
```

### Bipartite Graphs

- $G$ is bipartite if we can split $V$ into components $V_a$ and $V_b$ where each $e \in E$ is of the form <br>
$\{v \in V_a, v \in V_b\}$ <br>
- Denoted as $K_{m,\,n}$ <br>

<b>Example: </b><br>

$K_{2,\,3}$

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
G.add_nodes_from([1,2], bipartite=0) # Add the node attribute "bipartite"
G.add_nodes_from([3,4,5], bipartite=1)
G.add_edges_from([(1,3), (1,4), (1,5)
                ,(2,3), (2,4), (2,5)])

top = nx.bipartite.sets(G)[0]
pos = nx.bipartite_layout(G, top)
nx.draw(G, pos=pos, with_labels=True, node_color=['pink','magenta','red','deepskyblue','cyan'])
plt.show()
```

## Subgraphs

Graph $H$ is said to be a subgraph of Graph $G$ <br>
$IF:$ <br>
1. Vertices of $H$ is also a vertex of $G$ <br>
    $(V_H \subseteq V_G)$

2. Edges of $H$ is also an edges of $G$ <br>
    $(E_H \subseteq E_G)$

3. Edges of $H$ are <b><i>incident</i></b> to vertices of $G$ <br>

<b>Example: </b>


```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
H = nx.Graph()
G.add_edges_from([('A','B'),('A','D'),('D','C'),('B','C'),('A','C'),('B','D')])
H.add_edges_from([('A','B'),('B','C'),('C','A')])

# Manually specify node positions to form a diamond shape
pos = {'A': (0, 0), 'B': (-1, 1), 'C': (0, 2), 'D': (1, 1)}

plt.figure(figsize=(10, 5))  # Create a figure with a specific size

plt.subplot(1, 2, 1)  # Subplot 1 for the first graph
nx.draw_networkx_nodes(G, pos, node_size=500)
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color='black')
nx.draw_networkx_labels(G, pos)
plt.title('Graph G')
plt.axis('off')  # Turn off the axis

plt.subplot(1, 2, 2)  # Subplot 2 for the second graph
nx.draw_networkx_nodes(H, pos, node_size=500)
nx.draw_networkx_edges(H, pos, edgelist=H.edges(), edge_color='black')
nx.draw_networkx_labels(H, pos)
plt.title('Graph H')
plt.axis('off')  # Turn off the axis

plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

## The Handshake Theorem

If $G$ is any graph, then the sum of the degrees of all the vertices of $G$ equals twice the number of edges of $G$. Specifically, if the vertices of $G$ are $v_1,v_2,...,v_n$, where $n$ is a nonnegative integer.<br><br>
The total degree of $G = deg(v_1) + deg(v_2 ) +    + deg(v_n )$ = $2 \, \cdotp \textbf{(the number of edges of G)}$.<br>

<b>Example:</b><br>

G is a simple graph with 24 edges and a degree of 4 for each vertex. Determine the total number of vertices.

<b>Solution:</b><br>

$\textbf{The number of edges} = 24$

$\textbf{Each vertex's degree} = 4$

$\textbf{Let n be the total number of vertices in the graph.}$

$\textbf{Sum of the degree of all vertices = 2 x Number of edges}$ 

$n \cdot 4 = 2 \cdot 24 \, \textbf{by substituting the variables.}$

$4n = 48$

$n = 12$

$\textbf{Therefore, the total number of vertices in the graph is 12}$

## Travelling in a Graph

<b>Walks</b> - are a sequence of vertices and edges <br>

$W$ = $X \,e_1\, v_1\, e_2\, v _2\, e_3\, .\, .\, .\, v_{n - 1}\, e_n\, Y$ <br>

Where $e_i = \{X_{i - 1}, X_i\}$ for $1 \leq i \leq n$ 

<b>Closed Walks</b> - occurs when $X = Y$

<b>Trail</b> - is a walk with <b>NO</b> repeated edges <br><br>
<b>Path</b> - is a walk with <b>NO</b> repeated vertices <br><br>
<b>Closed Walk</b> - is a walk  from $v$ to $w$ that starts and ends in the same vertex.<br><br>
<b>Circuit</b> - is a closed walk that contains at least one edge and does not contain a repeated edge. <br><br>

## Euler's Circuit

- Traverses every edge once
- Every vertex has even degree

<b>Example: </b>

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a new graph
G = nx.Graph()
# Add nodes
for node in ['A', 'B', 'C', 'D']:
    G.add_node(node)

# Add edges between nodes
edges = [('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'A')]

# Assuming all nodes are interconnected as per the complex relationships
for edge in edges:
    G.add_edge(*edge)

# Draw the network
nx.draw(G,with_labels=True, node_color='lightblue', edge_color='gray')
plt.show()
```

<b>Solution: </b><br>
$\textbf{A, B, C, D, A}$ <br>


## Hamiltonian Circuits
